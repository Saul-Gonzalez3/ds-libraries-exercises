```python
# Create the users DataFrame.

users = pd.DataFrame({
    'id': [1, 2, 3, 4, 5, 6],
    'name': ['bob', 'joe', 'sally', 'adam', 'jane', 'mike'],
    'role_id': [1, 2, 3, 3, np.nan, np.nan]
})
users
```

```python
# Create the roles DataFrame

roles = pd.DataFrame({
    'id': [1, 2, 3, 4],
    'name': ['admin', 'author', 'reviewer', 'commenter']
})
roles
```

```python
(users.merge(roles, 
            left_on='role_id', 
            right_on='id', 
            how='outer')
    .drop(columns='role_id')
    .rename(columns={'id_x': 'id', 
                     'name_x': 'employee',
                     'id_y': 'role_id',
                     'name_y': 'role'}
            )
)
```

```python
#Load the mpg dataset from PyDataset.

from pydataset import data
mpg = data('mpg')
```

```python
#Output and read the documentation for the mpg dataset.

data('mpg', show_doc=True)
```

```python
#How many rows and columns are in the dataset?

mpg.shape
```

```python
#Check out your column names and perform any cleanup you may want on them.

mpg.describe()
mpg = mpg.rename(columns={'displ': 'display',
                    'cyl': 'cylinder',
                    'cty': 'city',
                     'hwy': 'highway',
                     'trans': 'transmission',
                      'drv': 'drive'})
mpg
```

```python
#Display the summary statistics for the dataset.

mpg.describe()
```

```python
#How many different manufacturers are there?


mpg.manufacturer.nunique()
```

```python
#How many different models are there?

mpg.model.nunique()
```

```python
#Create a column named mileage_difference like you did in the DataFrames exercises; this column should contain the difference between highway and city mileage for each car.

mpg['mileage_difference'] = mpg['highway'] - mpg['city']
mpg
```

```python
#Create a column named average_mileage like you did in the DataFrames exercises; this is the mean of the city and highway mileage.

mpg['average_mileage'] = (mpg['highway'] + mpg['city']) / 2
mpg
```

```python
#Create a new column on the mpg dataset named is_automatic that holds boolean values denoting whether the car has an automatic transmission.

mpg['is_automatic'] = np.where(mpg.transmission.isin(['auto(l5)','auto(av)','auto(s6)']), 'Automatic', 'Manual') 
mpg
```

```python
#Using the mpg dataset, find out which which manufacturer has the best miles per gallon on average?

mpg['average_mileage']
#mpg[mpg['average_mileage'] == mpg['average_mileage'].max()].manufacturer

mpg.groupby('manufacturer').mean().average_mileage.sort_values().tail(1)
```

```python
#Do automatic or manual cars have better miles per gallon?

#this might be an agg

mpg.groupby('is_automatic').average_mileage.agg(['mean']).round(2)
```

```python
#Use your get_db_url function to help you explore the data from the chipotle database.

url = env.get_db_url('chipotle')

chipotle_df = pd.read_sql('select * from chipotle.orders', url)
chipotle_df
```

```python
#What is the total price for each order?

chipotle_df['item_price'] = chipotle_df.item_price.str.strip('$').astype(float)

chipotle_df.groupby('order_id').item_price.sum()
```

```python
#What are the most popular 3 items?

chipotle_df.groupby('item_name').quantity.sum().sort_values(ascending=False).head(3)
```

```python
#Which item has produced the most revenue?

chipotle_df.groupby('item_name').item_price.sum().nlargest(1)
```
