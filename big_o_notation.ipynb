```python
#visualization imports
import matplotlib.pyplot as plt  

#standard DS imports
import numpy as np
import pandas as pd

#numerical imports
import math
from random import randint
import random
from math import log
```

```python
#Title your chart "Big O Notation"
#Label your x axis "Elements"
#Label your y axis "Operations"
#Label your curves or make a legend for the curves
#Use LaTex notation where possible

plt.figure(figsize=(8,8))

plt.xlim(1, 11)
plt.ylim(0, 50)

plt.suptitle('Big O Notation')
plt.xlabel('Elements')
plt.ylabel('Operations')


x = np.arange(0,20)
x2 = np.arange(0.01, 20)
y1 = [(0 * n + 1) for n in x]
y2 = [math.log(n) for n in x2]
y3 = [n for n in x]
y4 = [n * math.log(n) for n in x2]
y5 = [n**2 for n in x]
y6 = [2**n for n in x]
y7 = [np.math.factorial(n) for n in x]
y8 = [n**n for n in x]


plt.plot(x, y1, color='r', ls='--', label='$O(1)$')

plt.plot(x, y2, color='g', ls='--', marker='^', label='$O(log n)$')

plt.plot(x, y3, color='orange', ls='-.', marker='o', label='$O(n)$')

plt.plot(x, y4, color='blue', alpha=0.9, label='$O(n log n)$')

plt.plot(x, y5, color='r', ls='--', label='$O(n^2)$')

plt.plot(x, y6, color='g', ls='--', marker='^', label='$O(2^n)$')

plt.plot(x, y7, color='orange', ls='-.', marker='o', label='$O(n!)$')

plt.plot(x, y8, color='blue', label='$O(n^n)$')

plt.legend(loc='upper center') #You have to put labels with this


plt.show()
```
