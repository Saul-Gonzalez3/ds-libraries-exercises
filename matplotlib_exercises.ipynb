```python
#Use matplotlib to plot the following equation:


#You'll need to write the code that generates the x and y points.

#Add an anotation for the point 0, 0, as the origin.

x = range(0, 100)
y = [(n**2 - n + 2) for n in x]

plt.plot(x, y, marker = '*')
plt.title('$y=x^2 - x + 2$')
plt.xlabel('$x$')
plt.ylabel('$f(x)$')
plt.show()


```

```python
x = range(0, 120)
y = [1/(n+1) for n in x]

plt.barh(x, y)
plt.title('$y=1/(x+1)$')
plt.xlabel('$x$')
plt.ylabel('$f(x)$')
plt.show()
```

```python
#Combine the figures you created in the last step into one large figure with 4 subplots.

x = range(100)
y1 = [math.sqrt(n) for n in x]
x2 = range(100)
y2 = [n*n*n for n in x]
x3 = range(100)
y3 = [2**n for n in x]
x4 = range(100)
y4 = [1/(n+1) for n in x]


plt.figure(figsize=(8,8))
plt.suptitle('One Large Figure with 4 Subplots')

plt.subplot(221)
plt.plot(x, y1, color='r', ls='--')
plt.title('$y=sqrt(x)$')

plt.subplot(222)
plt.plot(x, y2, color='g', ls='--', marker='^')
plt.title('$y=x^3$')

plt.subplot(223)
plt.plot(x, y3, color='orange', ls='-.', marker='o')
plt.title('$y=2^x$')

plt.subplot(224)
plt.plot(x, y4, color='blue', alpha=0.9)
plt.title('$y=1/(x+1)$')

#plt.tight_layout()

plt.show()
```

```python
#Combine the figures you created in the last step into one figure where each of the 4 equations has a different color for the points. Be sure to include a legend and an appropriate title for the figure.


x = range(100)
y1 = [math.sqrt(n) for n in x]
x2 = range(100)
y2 = [n*n*n for n in x]
x3 = range(100)
y3 = [2**n for n in x]
x4 = range(100)
y4 = [1/(n+1) for n in x]


plt.figure(figsize=(8,8))
plt.suptitle('One Large Figure with 4 Subplots')

plt.subplot(221)
plt.plot(x, y1, color='r', ls='--', label='$y=sqrt(x)$')
plt.title('$y=sqrt(x)$')

plt.subplot(222)
plt.plot(x, y2, color='g', ls='--', marker='^', label='$y=x^3$')
plt.title('$y=x^3$')

plt.subplot(223)
plt.plot(x, y3, color='orange', ls='-.', marker='o', label='$y=2^x$')
plt.title('$y=2^x$')

plt.subplot(224)
plt.plot(x, y4, color='blue', alpha=0.9, label='$y=1/(x+1)$')
plt.title('$y=1/(x+1)$')

plt.legend(loc= 'lower left') #You have to put labels with this


plt.show()
```
