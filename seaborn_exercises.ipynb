```python
#Use seaborn's load_dataset function to load the iris database to answer the following questions:

iris = sns.load_dataset('iris')
iris.head()
```

```python
#What does the distribution of petal lengths look like?

sns.displot(data = iris, x = "petal_length", kde=True)
```

```python
#Is there a relationship between petal length and petal width?
sns.pairplot(iris, hue= ('petal_length'))
```

```python
#Would it be reasonable to predict species based on sepal width and sepal length? 
#For this, you'll visualize two numeric columns through the lense of a categorical column.

sns.catplot(data = iris, y = 'sepal_width', x = 'sepal_length', hue = 'species', kind = 'bar', errorbar = None)
```

```python
#Which features would be best used to predict species?
sns.relplot(x='species', y='petal_length', data=iris, hue="petal_length") #Winner Winner
sns.relplot(x='species', y='petal_width', data=iris, hue="petal_width")
sns.relplot(x='species', y='sepal_width', data=iris, hue="sepal_width")
sns.relplot(x='species', y='sepal_length', data=iris, hue="sepal_length")
```

```python
#Load the chipotle dataset from SQL, create a bar chart that shows the 4 most popular items and the revenue produced by each.

import env
url = env.get_db_url('chipotle')
orders = pd.read_sql('select * from orders', url)
orders.head()
```

```python
orders.item_price = orders.item_price.str.strip('$').astype(float)
```

```python
popular_items = orders.groupby('item_name').quantity.sum().nlargest(4)
popular_items
```

```python
popular_items_revenue = orders.groupby('item_name').item_price.sum().nlargest(4)
popular_items_revenue
```

```python
#Load the sleepstudy dataset from pydataset and read it's documentation. 

sleepstudy = data('sleepstudy')
data('sleepstudy', show_doc=True)
```

```python
#Use seaborn to create a line chart of all the individual subject's reaction times and a more prominant line showing the average change in reaction time.
```

```python
plt.figure(figsize=(12,6))
sns.lineplot(data=sleepstudy, x='Days', y='Reaction', hue='Subject')
plt.show()

sleepstudy.Subject = sleepstudy.Subject.astype(str)


plt.figure(figsize=(12,8))

sns.lineplot(data=sleepstudy, x='Days', y='Reaction', hue='Subject', alpha=.5)
sns.lineplot(data=sleepstudy, x='Days', y='Reaction', 
             color = 'black', errorbar=('ci',False), linewidth=3)

plt.title('Reaction Time of Subjects Over Time')
plt.legend(loc='upper right')
plt.show()
```
